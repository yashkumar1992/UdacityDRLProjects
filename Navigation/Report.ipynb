## Learning Algorithm

The Algorithm used to train the Agent here is called "Deep Q Networks".
Deep Q networks computes the Action to be taken in a given environmental conditions like current state, reward.
Deep Q Network is powerful than the conventional Tabular Methods because:  
    1. Tabular Methods requires a huge storage space as State,Action Space grows.
    2. Ability to Generalize through Function Approximation.
    
### Deep Q Network -  Neural Network Architecture
we are using 2 relu layers here. We can increase/decrease layers based on the complexity of the task.
##### Parameters :   
state_size (int): Dimension of each state  
action_size (int): Dimension of each action  
seed (int): Random seed  
fc1_units (int): Number of nodes in first hidden layer  
fc2_units (int): Number of nodes in second hidden layer


### Deep Q Network -  Agent
We use the above mentioned architecture to train the Agent in Learning a Task by sending reward signals.
Agent has following Functionalities
##### Act:
This the Agent's Policy to choose an Action. We can use Softmax, Epsilon Greedy for this. We are using Epsilon Greedy.  Epsilon - Greedy - works by randomizing exploration n% of the times thereby learning from those explorations.

##### Learn:
This Module trains the local network by updating the weights by backpropogation that is governed by Temporal Difference Learning. 
Deep Q Networks overcomes problems in training by:  
1. To avoid the cyclical trap of predictions chasing itself(leading to no convergence) -> 2 identical networks are used. One to Predict at every step but updated once in a while. Other to update at every step.This is called Target - Q Network
2. To avoid the sequential correlation between actions, we need a database to collect the experience and replay them by sampling whenever needed. This is acheived by Experience Replay.

Finally the learned weights are updated by soft-update so that similar states values are updated.

The Exploration/Exploitation ratio diminishes over time, thereby converging to give the most profitable actions.
This is acheived by decaying the epsilon value.

Agent is trained with following parameters:  
#### Parameters:
n_episodes=2000, max_t=1000, eps_start=1.0, eps_end=0.01, eps_decay=0.995


#### Parameters
replay buffer size  = int(1e5)
minibatch size = 64
discount factor = 0.99 
for soft update of target parameters = 1e-3
learning rate = 5e-4 
how often to update the network  = 10
